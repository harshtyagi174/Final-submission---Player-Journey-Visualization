# Streamlit Environment Debugging

This notebook checks the local Python environment for Streamlit installation, writes a minimal dashboard app, and attempts to run it.

```python
import sys
import subprocess
from pathlib import Path

print('Python executable:', sys.executable)
print('Python version:', sys.version)
print('Current working directory:', Path.cwd())

print('\n-- pip show streamlit --version --packages --find-links --no-python-version-warning --disable-pip-version-check')
subprocess.run([sys.executable, '-m', 'pip', 'show', 'streamlit'])
print('\n-- importlib find_spec')
import importlib.util
print(importlib.util.find_spec('streamlit'))
```
